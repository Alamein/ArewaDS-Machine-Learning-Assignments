```python
import pandas as pd
import matplotlib.pyplot as plt
birds = pd.read_csv('../Week_3/data/Birds.csv')
birds.head()
```

```python
snow_goose_s = birds[birds['Name']=='Snow goose']
snow_goose_s.head()
```

```python
snow_goose_s.plot()
```

```python
family = snow_goose_s['Family']
Order = snow_goose_s['Order']
Genus = snow_goose_s['Genus']

plt.barh(Genus, family)
plt.barh(Genus, Order)

plt.show()
```

```python
snow_goose_s.plot('Name')
```

```python
minLength = birds[3:]['MinLength']
maxLength = birds[3:]['MaxLength']
category = birds[3:]['Category']
name = birds[3:]['Name']
plt.barh(name, maxLength)
plt.barh(name, minLength)
plt.barh(category, maxLength)
plt.barh(category, minLength)
plt.rcParams['figure.figsize'] = [6,12]

plt.show()
```
