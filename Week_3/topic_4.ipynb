```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

honey = pd.read_csv('../Week_3/data/honey.csv')
honey.head()
```

```python
# line
sns.relplot(data=honey,
            x='year',y='stocks',
            col='state',
            col_wrap=3,
            kind='line')
```

```python
sns.relplot(data=honey,
            x='priceperlb',y='numcol',
            col='year',
            col_wrap=3,
            kind='line')
```

```python
fig, ax = plt.subplots(figsize=(12,6))
lineplot = sns.lineplot(x=honey['year'], y=honey['numcol'], data=honey, 
                        label = 'Number of bee colonies', legend=False)
sns.despine()
plt.ylabel('# colonies')
plt.title('Price vs number of bee colonies over the years');

ax2 = ax.twinx()
lineplot2 = sns.lineplot(x=honey['year'], y=honey['priceperlb'], ax=ax2, color="r", 
                         label ='Price per lb', legend=False) 
sns.despine(right=False)
plt.ylabel('price per lb')
ax.figure.legend();
```

```python
fig, ax = plt.subplots(figsize=(12,6))
lineplot = sns.lineplot(x=honey['year'], y=honey['priceperlb'], data=honey, 
                        label = 'Price per lb', legend=False)
sns.despine()
plt.ylabel('# Prices')
plt.title('Price vs stock over the years');

ax2 = ax.twinx()
lineplot2 = sns.lineplot(x=honey['year'], y=honey['stocks'], ax=ax2, color="r", 
                         label ='stocks', legend=False) 
sns.despine(right=False)
plt.ylabel('stocks')
ax.figure.legend();
```
